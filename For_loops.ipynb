```python
string = 'dataabcs!$!$%!%%4science'
for i in string:
  if i.isalpha()==False:
    print(i, end=',')
```

```python
string = 'dataabcs!$!$%!%%4science'
for i in string:
  if not i.isalpha():
    print(i, end=',')
```

```python
string = 'dataabcs!$!$%!%%4science'
counter = 0
for i in string:
  if not i.isalpha():
    counter = counter + 1
    print(counter, end = " ")
```
